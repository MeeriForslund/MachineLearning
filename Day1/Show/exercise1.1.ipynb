```python
import numpy as np
import matplotlib.pyplot as plt


def y(x):
    return k*(0.25)*x

# x = np.array([1,5,9,11])
# or use np.linspace(0,11,12)
x = np.linspace(-1,11,12)
colors = ['r','b','k','g']
for i in range(1,4):
    k = i*(0.25)
    plt.plot(x,y(x), color=colors[i])



plt.title('y=kx')
plt.xlabel('x-value')
plt.ylabel('y-value')
plt.show
```

Let's do m in $ y = kx+m $

```python
k = 0.5

m = np.linspace(0.25,0.75,3)

def y2(m):
    return k*x+m
i=1
for ms in m:
    plt.plot(x,y2(ms), color=colors[i])
    i=i+1

plt.title('y=kx+m')
plt.xlabel('x-value')
plt.ylabel('y-value')
plt.show
```
