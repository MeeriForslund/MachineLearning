```python
import numpy as np
import matplotlib.pyplot as plt


def y(x):
    return k*(0.25)*x

x = np.array([1,5,9,11])

colors = ['r','b','k','g']
for i in range(1,4):
    k = i*(0.25)
    plt.plot(x,y(x), color=colors[i])



plt.title('y=kx')
plt.xlabel('x-value')
plt.ylabel('y-value')
plt.show
```
